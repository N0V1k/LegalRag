```python
import json
```

```python
with open('ТК.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# вывод структуры файла
def print_nested_structure(data, level=1, path=None, counter=None):
    
    if path is None:
        path = []
    if counter is None:
        counter = [1] 
    
    indent = "    " * (level - 1)
    
    if isinstance(data, dict):
        for key, value in data.items():
            print(f"{indent}├── {key}")
            print_nested_structure(value, level + 1, path + [key], counter)
    elif isinstance(data, list):
        for i, item in enumerate(data, 1):
            if i == 1:
                counter[0] = 1
            print(f"{indent}├── [{counter[0]}] {item}")
            counter[0] += 1
    else:
        
        # с текстами статей
        #print(f"{indent}├── {data}")
        
        # без текстов статей
        print(f"{indent}├── Текст статьи")
        
        counter[0] += 1
```

```python
print_nested_structure(data)
```
